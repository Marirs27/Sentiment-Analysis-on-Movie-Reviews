```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.metrics import f1_score
import numpy as np
from sklearn.metrics import classification_report
from sklearn.metrics import plot_confusion_matrix
```

```python
df = pd.read_csv("clean_reviews.csv")
X = df["preprocessed_text"]
y = df["sentiment"]
```

```python
Count_vec = CountVectorizer(max_df=0.90, min_df=2, max_features=1000, stop_words='english')
X_bow = Count_vec.fit_transform(X)
X_train,X_test,y_train,y_test = train_test_split(X_bow,y,test_size=0.2,random_state=42)
```

## Logical Regression

```python
from sklearn.linear_model import LogisticRegression

regressor = LogisticRegression(max_iter=200)
regressor.fit(X_train,y_train)
predicted = regressor.predict(X_test)
print(classification_report(y_test,predicted,labels=[0,1]))
plot_confusion_matrix(regressor, X_test, y_test,cmap=plt.cm.Blues)  
plt.show()
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier

for i in [1,3,5,7,9,11] :
    neigh = KNeighborsClassifier(n_neighbors=i)
    neigh.fit(X_train,y_train)
    predicted = neigh.predict(X_test)
    print(classification_report(y_test,predicted,labels=[0,1]))
    plot_confusion_matrix(neigh, X_test, y_test,cmap=plt.cm.Blues)  
    plt.show()
```

## Support Vector Machine

```python
from sklearn import svm

classifier = svm.SVC(kernel='poly')
classifier.fit(X_train,y_train)
predicted = classifier.predict(X_test)
print(classification_report(y_test,predicted,labels=[0,1]))
plot_confusion_matrix(classifier, X_test, y_test,cmap=plt.cm.Blues)  
plt.show()
```

## Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB

classifier = MultinomialNB()
classifier.fit(X_train,y_train)
predicted = classifier.predict(X_test)
print(classification_report(y_test,predicted,labels=[0,1]))
plot_confusion_matrix(classifier,X_test,y_test,cmap=plt.cm.Blues)
plt.show()
```
